```python
bank_m=pd.read_csv('bank.csv',sep=';')
```

>***Rename Columns***

```python
bank_m=bank_m.rename(columns={'age':'age_of_client','job':'type_of_job','marital':'marital_status','education':'education_types',\
                      'default':'has_credict_or_not','balance':'balance','contact':'communication_type',\
                     'day':'day_of_week','month':'month_of_the_year','duration':'last_contact_duration',\
                     'campaign':"number_of_contacts",'pdays':'previous_days_of_contact',
                      'previous':'no_of_contacts_before_campaign','poutcome':'outcome_of_previous_campaign',\
                     'y':'target'})
```

```python
import seaborn as sns
vis3=sns.barplot(bank_m['target'],bank_m['balance'])
```

***Remarks***
<br>
It is observed that clients who have more balance are prone to subscribe to fixed deposit programme.
